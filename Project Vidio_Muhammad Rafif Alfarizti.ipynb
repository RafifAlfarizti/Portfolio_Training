## Mengatasi outlier

```python
# Fungsi untuk menghapus outliers menggunakan metode IQR
def remove_outliers_iqr(vid):
    # Menyimpan indeks outliers
    outliers_indices = set()
    
    # Hanya memproses kolom numerik
    for col in vid.select_dtypes(include=[np.number]).columns:
        # Menghitung Q1, Q3, dan IQR
        Q1 = np.percentile(vid[col].dropna(), 25)
        Q3 = np.percentile(vid[col].dropna(), 75)
        IQR = Q3 - Q1
        
        # Menentukan batas bawah dan atas
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Menyimpan indeks dari nilai outliers
        outliers_indices.update(vid[(vid[col] < lower_bound) | (vid[col] > upper_bound)].index)
    
    # Menghapus baris yang mengandung outliers
    vid_clean = vid.drop(outliers_indices)
    return vid_clean

# Menghapus baris yang mengandung outliers
vid_clean = remove_outliers_iqr(vid)
print("Dataset tanpa outlier:")
print(vid_clean)
```
